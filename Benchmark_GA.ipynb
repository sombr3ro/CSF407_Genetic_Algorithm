```python
import Improved_GA
import Naive_GA
from CNF_Creator import CNF_Creator
import matplotlib.pyplot as plt
```

```python
average_fitness_iGA = {}
average_time_iGA = {}

for m in range(100,301,20):
    avg_fit = 0.0
    avg_time = 0.0
    run = 10
    for k in range(run):
        cnf_stmt = CNF_Creator(50).CreateRandomSentence(m)
        fitness, _, time_taken = Improved_GA.run_algo(cnf_stmt)
        avg_fit += fitness
        avg_time += time_taken
    average_fitness_iGA[m] = avg_fit/float(run)
    average_time_iGA[m] = avg_time/float(run)

print(average_fitness_iGA)
print(average_time_iGA)
```

```python
average_fit_naiveGA = {}
average_time_naiveGA = {}

for m in range(100,301,20):
    avg_fit = 0.0
    avg_time = 0.0
    run = 10
    for k in range(run):
        cnf_stmt = CNF_Creator(50).CreateRandomSentence(m)
        fitness, _, time_taken = Naive_GA.run_algo(cnf_stmt)
        avg_fit += fitness
        avg_time += time_taken
    average_fit_naiveGA[m] = avg_fit/float(run)
    average_time_naiveGA[m] = avg_time/float(run)

print(average_fit_naiveGA)
print(average_time_naiveGA)
```

```python
plt.figure(figsize=(10,5))
plt.plot(average_fitness_iGA.keys(), average_fitness_iGA.values())
plt.title("Average Fitness vs No. of Clauses")
plt.style.use('ggplot')
plt.plot(average_fitness_iGA.keys(), average_fitness_iGA.values(), 'o-b')
plt.xlabel("Number of Clauses")
plt.ylabel("Fitness score (%)")
plt.grid(True, which= "both")
```

```python
plt.figure(figsize=(10,5))
plt.plot(average_fitness_iGA.keys(), average_fitness_iGA.values())
plt.title("Average Time Taken vs No. of Clauses")
plt.style.use('ggplot')
plt.plot(average_time_iGA.keys(), average_time_iGA.values(), 'o-b')
plt.xlabel("Number of Clauses")
plt.ylabel("Average Time taken (s)")
plt.ylim(ymin=0, ymax = 40)
plt.grid(True, which= "both")
```

```python
plt.figure(figsize=(10,5))
plt.title("Average Fitness vs No. of Clauses")
plt.style.use('ggplot')
plt.plot(average_fitness_iGA.keys(), average_fitness_iGA.values(), 'o-b', label = 'Improved Genetic Algo')
plt.plot(average_fit_naiveGA.keys(), average_fit_naiveGA.values(), 'o-r', label = 'Naive Genetic Algo')
plt.legend()
plt.xlabel("Number of Clauses")
plt.ylabel("Fitness score (%)")
plt.grid(True, which= "both")
```

```python
plt.figure(figsize=(10,5))
plt.title("Average Time Taken vs No. of Clauses")
plt.style.use('ggplot')
plt.plot(average_time_iGA.keys(), average_time_iGA.values(), 'o-b', label = 'Improved Genetic Algo')
plt.plot(average_time_naiveGA.keys(), average_time_naiveGA.values(), 'o-r', label = 'Naive Genetic Algo')
plt.xlabel("Number of Clauses")
plt.ylabel("Average Time taken (s)")
plt.legend()
plt.ylim(ymin=0)
plt.grid(True, which= "both")
```
